```python
import numpy as np
import scipy
import matplotlib
import matplotlib.pyplot as plt
```

**Zadanie 3**

```python
# 1.
x1 = np.linspace(-1, 1, 50)
plt.plot(x1, x1**2 +5)
plt.grid()
plt.ylabel("y")
plt.xlabel('x')
plt.title("y = x^2 + 5 dla -1<x<1")
plt.legend("y")
```

```python
# 2. 
x1 = np.linspace(-6, 6, 150)
plt.plot(x1, x1**2 +5, 'g')
plt.grid()
plt.ylabel("y")
plt.xlabel('x')
plt.title("y = x^2 + 5 dla -6<x<6")
plt.legend("y")
```

```python
# 3. 
x1 = np.linspace(0, 5, 150)
plt.plot(x1, x1**2 +5, 'r')
plt.grid()
plt.ylabel("y")
plt.xlabel('x')
plt.title("y = x^2 + 5 dla 0<x<5")
plt.legend("y")
```
