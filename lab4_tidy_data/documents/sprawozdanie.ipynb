```python
import pandas as pd
data = pd.read_csv('../Analysis data/earthquake_data.csv')
```

Zmiana nazw kolumn i usunięcie rzędów, które nie zostały w całości wypełnione.

```python
data.rename(columns = {'What is your gender?': 'Gender', 
                       'Do you think the "Big One" will occur in your lifetime?': 'Belief in Big One occuring in their lifetime',
                       'In general, how worried are you about earthquakes?': 'Worry about earthquakes',
                       'How worried are you about the Big One, a massive, catastrophic earthquake?': 'Worry about the Big One',
                       'Have you ever experienced an earthquake?': 'Experience of an earthquake',
                       'Have you or anyone in your household taken any precautions for an earthquake (packed an earthquake survival kit, prepared an evacuation plan, etc.)?': 'Preacautions were taken',
                       'How familiar are you with the San Andreas Fault line?': 'Familiarity with San Andreas Fault line',
                       'How familiar are you with the Yellowstone Supervolcano?': 'Familiarity with Yellowstone Volcano',
                       'How much total combined money did all members of your HOUSEHOLD earn last year?': 'Annual income for household'}, inplace = True)
data = data.dropna()
```
